```python
import sys
sys.path.append('./')
from datasets.dataloader import load_cifar10, RotatedDataset
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
from copy import deepcopy
```

```python
trainset, testset = load_cifar10()
rotated_trainset = RotatedDataset(deepcopy(trainset), 90)
rotated_testset = RotatedDataset(deepcopy(testset), 90)
```

```python
normal_trainloader = DataLoader(trainset, batch_size=1, shuffle=False)
rotated_trainloader = DataLoader(rotated_trainset, batch_size=1, shuffle=False)
fig, axs = plt.subplots(3, 2, figsize=(10, 15))

for i, batch in enumerate(normal_trainloader):
    if i == 3:
        break
    normal_img = batch[0].squeeze(0)
    axs[i,0].imshow(normal_img.numpy().transpose(1, 2, 0))
    axs[i,0].set_title(f'Normal Image {i}')

for i, batch in enumerate(rotated_trainloader):
    if i == 3:
        break
    rotated_img = batch[0].squeeze(0)
    axs[i,1].imshow(rotated_img.numpy().transpose(1, 2, 0))
    axs[i,1].set_title(f'Rotated Image {i}')
```
